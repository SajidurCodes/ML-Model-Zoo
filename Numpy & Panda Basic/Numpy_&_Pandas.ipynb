```python
# Perform various basic and cumulative aggregations on a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Basic Aggregations
print("Sum:", np.sum(arr))
print("Product:", np.prod(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
print("Variance:", np.var(arr))
# Variance measures how much the values in a dataset deviate from the mean.
#A higher variance means the data points are more spread out.

# Min & Max Functions
print("Min:", np.min(arr))
print("Max:", np.max(arr))
print("Index of Min:", np.argmin(arr))
print("Index of Max:", np.argmax(arr))

# Cumulative Aggregations
print("Cumulative Sum:", np.cumsum(arr))
print("Cumulative Product:", np.cumprod(arr))

# Percentile & Quantile
print("25th Percentile:", np.percentile(arr, 25))
print("50th Percentile (Median):", np.percentile(arr, 50))
print("Quantile 0.5:", np.quantile(arr, 0.5))

# Logical Aggregations
bool_arr = np.array([True, False, True])
print("All True?", np.all(bool_arr))
print("Any True?", np.any(bool_arr))

# Other Statistical Functions
print("Range (Max - Min):", np.ptp(arr))
print("Correlation Coefficient:\n", np.corrcoef(arr))
print("Covariance Matrix:\n", np.cov(arr))
```

```python
# Perform various aggregations on a 2D NumPy array with NaN values, ignoring NaNs
# 2D matrix with NaN values
arr = np.array([
    [1, 2, np.nan],
    [4, 5, 6],
    [7, 8, 9]
])

# Basic Aggregations
print("Sum:", np.nansum(arr))  # Ignores NaN
print("Product:", np.nanprod(arr))  # Ignores NaN
print("Mean:", np.nanmean(arr))  # Ignores NaN
print("Median:", np.nanmedian(arr))  # Ignores NaN
print("Standard Deviation:", np.nanstd(arr))  # Ignores NaN
print("Variance:", np.nanvar(arr))  # Ignores NaN

# Min & Max Functions
print("Min:", np.nanmin(arr))
print("Max:", np.nanmax(arr))
print("Index of Min:", np.nanargmin(arr))
print("Index of Max:", np.nanargmax(arr))

# Cumulative Aggregations
print("Cumulative Sum:\n", np.nancumsum(arr))
print("Cumulative Product:\n", np.nancumprod(arr))

# Percentile & Quantile
print("25th Percentile:", np.nanpercentile(arr, 25))
print("50th Percentile (Median):", np.nanpercentile(arr, 50))
print("Quantile 0.5:", np.nanquantile(arr, 0.5))

# Logical Aggregations
bool_arr = np.array([[True, False, True], [False, True, True]])
print("All True?", np.all(bool_arr))
print("Any True?", np.any(bool_arr))

# Other Statistical Functions
print("Range (Max - Min):", np.ptp(arr))
print("Correlation Coefficient:\n", np.corrcoef(np.nan_to_num(arr)))  # Replace NaN with 0 for correlation
print("Covariance Matrix:\n", np.cov(np.nan_to_num(arr), rowvar=False))  # Replace NaN with 0 for covariance
```

```python
# Demonstrate advanced indexing techniques for NumPy arrays, including integer array indexing, boolean masking, and assigning values
arr = np.array([10, 20, 30, 40, 50])
indices = [0, 2, 4]  # Selecting elements at indices 0, 2, and 4

print(arr[indices])  # Output: [10 30 50]


arr2d = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

row_indices = [0, 2]  # Select rows 0 and 2
col_indices = [1, 2]  # Select columns 1 and 2

print(arr2d[row_indices])  # Selects rows 0 and 2
print(arr2d[:, col_indices])  # Selects columns 1 and 2


print(arr2d[[0, 2], [1, 2]])  # Selects elements (0,1) and (2,2) -> [20 90]


mask = np.array([True, False, True, False, True])  # Select where True
print(arr[mask])  # Output: [10 30 50]


arr[indices] = [100, 300, 500]
print(arr)  # Output: [100  20 300  40 500]
```

```python
# NumPy: dtype, astype, views vs copies

arr = np.arange(6, dtype=np.int64)
print("dtype:", arr.dtype, "itemsize:", arr.itemsize)

# convert dtype (creates copy)
arr_float = arr.astype(np.float32)
print("converted dtype:", arr_float.dtype)

# views vs copies
a = np.arange(10)
b = a.view()    # view shares memory
c = a.copy()    # independent
b[0] = 999
print("a after modifying b (view):", a[:3])
c[1] = -111
print("a after modifying c (copy):", a[:3])   # unchanged

# memory layout (C vs Fortran)
m = np.arange(12).reshape(3,4)
print("C-contiguous:", m.flags['C_CONTIGUOUS'], "F-contiguous:", m.flags['F_CONTIGUOUS'])
```

```python
# Broadcasting and conditional selection
a = np.array([0,1,2])
b = np.array([10])
print("a + b (broadcast):", a + b)

# conditional: replace negatives with 0
x = np.array([-2, 0, 3, -1, 5])
y = np.where(x < 0, 0, x)
print("negatives replaced:", y)

# multiple conditions with select
conds = [x < 0, (x >= 0) & (x < 3), x >= 3]
choices = ["neg", "small", "large"]
labels = np.select(conds, choices, default="unknown")
print(labels)
```

```python
# Linear algebra with numpy
A = np.array([[3., 1.], [1., 2.]])
b = np.array([9., 8.])

# solve Ax = b
x = np.linalg.solve(A, b)
print("solution x:", x)

# inverse (not recommended in practice to compute solve via explicit inverse)
A_inv = np.linalg.inv(A)
print("A_inv @ b:", A_inv @ b)

# eigenvalues / eigenvectors
w, v = np.linalg.eig(A)
print("eigenvalues:", w)
print("eigenvectors:\n", v)

# SVD
U, s, Vt = np.linalg.svd(A)
print("SVD singular values:", s)
```

```python
# Count and get proportions of unique values in 'type' and 'department' columns

store_counts = store_types["type"].value_counts()
print(store_counts)

# Get the proportion of stores of each type
store_props = store_types["type"].value_counts(normalize=True)
print(store_props)

# Count the number of stores of each department and sort
dept_counts_sorted = store_depts["department"].value_counts(sort=True)
print(dept_counts_sorted)

# Get the proportion of stores of each department and sort
dept_props_sorted = store_depts["department"].value_counts(sort=True, normalize=True)
print(dept_props_sorted)
```

```python
# Calculate total weekly sales and sales proportions for different store types
sales_all = sales["weekly_sales"].sum()

# Subset for type A stores, calc total weekly sales
sales_A = sales[sales["type"] == "A"]["weekly_sales"].sum()

# Subset for type B stores, calc total weekly sales
sales_B = sales[sales["type"] == "B"]["weekly_sales"].sum()

# Subset for type C stores, calc total weekly sales
sales_C = sales[sales["type"] == "C"]["weekly_sales"].sum()

# Get proportion for each type
sales_propn_by_type = [sales_A, sales_B, sales_C] / sales_all
print(sales_propn_by_type)
```

```python
# Create a pivot table to count how many subjects each student took
df.pivot_table(values='Score', index='Student', columns='Subject', aggfunc='count')
```

```python
# Create a pivot table to calculate the total score per subject
# Total Score per Subject

df.pivot_table(values='Score', index='Subject', aggfunc='sum')
```

```python
# Create a pivot table to calculate the average score by student and subject, filling NaN values with 0
# Use fill_value to Replace NaN

df.pivot_table(values='Score', index='Student', columns='Subject', aggfunc='mean', fill_value=0)
```

```python
# Create a pivot table with multiple aggregations (mean and max) for scores by student and subject
# Multiple Aggregations
df.pivot_table(values='Score', index='Student', columns='Subject', aggfunc=['mean', 'max'])
```

# **Vectorized String Operations**

What Does "Vectorized" Mean?
It means you can apply operations on entire columns of strings without using loops.
Fast, efficient, and clean!





```python
# Create a DataFrame with 'Name' and 'City' columns containing strings for vectorized string operations
df = pd.DataFrame({
    'Name': [' Alice ', 'BOB', 'charlie', 'D@vid'],
    'City': ['Dhaka', 'chittagong', 'SYLHET', 'Rajshahi']
})
```

```python
# Convert 'Name' to lowercase and 'City' to title case using vectorized string methods
df['Name_lower'] = df['Name'].str.lower()
df['City_title'] = df['City'].str.title()
print(df)
```

```python
# Remove leading/trailing whitespace from the 'Name' column using vectorized string methods
df['Name_clean'] = df['Name'].str.strip()
print(df)
```

```python
# Remove special characters from the 'Name' column using vectorized string methods and regex
df['Name_no_special'] = df['Name'].str.replace(r'[^a-zA-Z]', '', regex=True)
print(df)
```

```python
# Filter DataFrame rows where the 'City' column contains 'dha' (case-insensitive)
filtered = df[df['City'].str.contains('dha', case=False)]
print(df)
```

```python
# Filter DataFrame rows where the 'Name' column starts with 'A' or ends with 'd'
df[df['Name'].str.startswith('A')]
df[df['Name'].str.endswith('d')]
```

```python
# Add a new column 'Name_len' with the length of each string in the 'Name' column
df['Name_len'] = df['Name'].str.len()
print(df)
```

```python
# Extract the first alphanumeric character from the 'Name' column using regex
df['First_letter'] = df['Name'].str.extract(r'^(\w)')
print(df)
```

```python
# Apply a chain of vectorized string operations to clean the 'Name' column
df['Cleaned_Name'] = df['Name'].str.strip().str.lower().str.replace(r'[^a-z]', '', regex=True)
print(df)
```
